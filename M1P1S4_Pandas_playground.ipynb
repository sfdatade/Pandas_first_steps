```python
import pandas as pd

# Create a dictionary with the fruit information
fruit_dict = {
    'Fruit': ['Apple', 'Banana', 'Orange', 'Grape', 'Mango', 'Pineapple'],
    'Weight (oz)': [4.2, 6.1, 3.2, 1.6, 8.9, 13.1],
    'Color': ['Red', 'Yellow', 'Orange', 'Purple', 'Yellow', 'Yellow/Brown'],
    'Taste': ['Sweet', 'Sweet', 'Sour', 'Sweet', 'Sweet', 'Sweet/Tangy'],
    'Vitamin C (mg)': [8, 10, 70, 0.8, 46, 78]
}

# Create a pandas DataFrame from the dictionary
fruit_df = pd.DataFrame(fruit_dict)

# Print the DataFrame
print(fruit_df)

# This will create a DataFrame with 5 columns (Fruit, Weight (oz), Color, Taste, and Vitamin C (mg)) and 6 rows containing information about different kinds of fruits. The columns contain information such as the fruit's weight, color, taste, and vitamin C content, and the rows correspond to different types of fruit (Apple, Banana, Orange, Grape, Mango, and Pineapple).
```
